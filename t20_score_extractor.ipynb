```python
import pandas as pd
import numpy as np
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for files in os.listdir('t20s'):
    filenames.append(os.path.join('t20s',files))
```

```python
filenames[:5]
```

```python
final_df_list = []
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df_list.append(df)
        counter +=1

final_df = pd.concat(final_df_list, ignore_index = True)
```

```python
final_df
```

```python
final_df.drop(columns = [
    'meta.data_version',
    'meta.created',
    'meta.revision',
    'info.outcome.bowl_out',
    'info.bowl_out',
    'info.supersubs.South Africa',
    'info.supersubs.New Zealand',
    'info.outcome.eliminator',
    'info.outcome.result',
    'info.outcome.method',
    'info.neutral_venue',
    'info.match_type_number',
    'info.outcome.by.runs',
    'info.outcome.by.wickets',
], axis = 1, inplace = True)
```

```python
import pickle
matches = pickle.load(open('t20_dataset.pkl', 'rb'))
matches
```

```python
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
count = 1
delivery_df_list = []
for index,row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1    
    batsman = []
    bowler = []
    batting_team = []
    match_id = []
    ball_of_match = []
    teams = []
    city = []
    venue = []
    player_of_dismissed = []
    runs = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            ball_of_match.append(key)
            teams.append(row['info.teams'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            runs.append(ball[key]['runs']['total'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')

    loop_df = pd.DataFrame({
        'match_id' : match_id,
        'teams' : teams,
        'batting_team' : batting_team,
        'ball_of_match' : ball_of_match,
        'batsman' : batsman,
        'bowler' : bowler,
        'runs' : runs,
        'city' : city,
        'venue' : venue,
        'player_of_dismissed' : player_of_dismissed
})
    delivery_df_list.append(loop_df)
    delivery_df = pd.concat(delivery_df_list, ignore_index = True)
```

```python
import pickle

# pickle.dump(delivery_df,open('t20_dataset_1.pkl','wb'))
deliveries = pickle.load(open('t20_dataset_1.pkl', 'rb'))
deliveries
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
deliveries['bowling_team'] = deliveries.apply(bowl,axis = 1)
```

```python
deliveries
```

```python
deliveries.drop(columns  = 'teams', inplace = True)
deliveries
```

```python
teams = [
    'Australia',
    'India',
    'South Africa',
    'England',
    'Pakistan',
    'Afghanistan',
    'West Indies',
    'Sri Lanka',
    'New Zealand'
]
```

```python
deliveries = deliveries[deliveries['batting_team'].isin(teams)]
deliveries = deliveries[deliveries['bowling_team'].isin(teams)]

deliveries
```
