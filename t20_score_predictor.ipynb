```python
import pickle
import numpy as np
import pandas as pd
```

```python
df= pickle.load(open('t20_dataset_2.pkl', 'rb'))
```

```python
df
```

```python
df['city'].isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x : x[0]),df['city']) 
cities
```

```python
df['venue'].str.split()
```

```python
df['city'] = cities
df['city'].isnull().sum()
```

```python
df.drop(columns = ['venue'], inplace = True)
df
```

```python
df.rename(columns ={'city' : 'venue'},inplace = True )
```

```python
df
```

```python
new_venue = df['venue'].value_counts()[df['venue'].value_counts() > 600].index.tolist()
df = df[df['venue'].isin(new_venue)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
df
```

```python
df['over'] = df['ball_of_match'].apply(lambda x: str(x).split('.')[0])
df['ball_no'] = df['ball_of_match'].apply(lambda x: str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_of_dismissed'] = df['player_of_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_of_dismissed'] = df['player_of_dismissed'].astype('int')
df['player_of_dismissed'] = df.groupby('match_id')['player_of_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_of_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
df['match_id'].unique()
```

```python
groups = df.groupby('match_id')

last_five = []
match_id = df['match_id'].unique()
for id in match_id:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
last_five
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id')['runs'].sum().reset_index().merge(df,on='match_id')
```

```python
final_df
```

```python
final_df.rename(columns = {'runs_x': 'total_runs', 'runs_y':'runs'}, inplace = True)
```

```python
final_df = final_df[['batting_team', 'bowling_team', 'venue', 'current_score', 'balls_left', 'wickets_left', 'crr', 'last_five', 'total_runs']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final_df.drop(columns = 'total_runs')
y = final_df['total_runs']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop = 'first'),['batting_team', 'bowling_team', 'venue'])
    ]
    ,remainder = 'passthrough'
    )
```

```python
pipe = Pipeline(steps = [
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimator = 1000,learning_rate = 0.2, max_depth = 12, random_state = 1)),
])
```

```python
pipe.fit(x_train, y_train)
```

```python
x_train.shape
```

```python
y_pred = pipe.predict(x_test)
print("First 5 Predictions", y_pred[:5])
print("R2_score", r2_score(y_test,y_pred))
print("Mean absolute error", mean_absolute_error(y_test,y_pred))
```
